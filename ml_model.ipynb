# LOAD DATA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_pickle('armyloss_change.pkl')
```

# EDA

```python
fig = plt.figure()
sns.distplot(data['Manpower'], bins = 50)
```

```python
print("Skewness: %f" % data['Manpower'].skew())
print("Kurtosis: %f" % data['Manpower'].kurt())
```

```python
data[data['Manpower'] > 500]
```

```python
var = 'Date'
f, ax = plt.subplots(figsize=(16, 8))
fig = sns.scatterplot( x = var, y = 'Manpower', data = data)
```

## Descriptive Statistics

```python
data.describe()
```

## Correlation Matrix

```python
f,ax = plt.subplots(figsize=(15,10))
sns.heatmap(data.corr(), annot =True, linewidth =".5", fmt =".2f", cmap='GnBu')
plt.show()
```

## Missing Values

```python
sns.heatmap(data.isnull(), cbar = False)
```
